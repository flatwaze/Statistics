1. Построить на одном графике графики плотностей N(0, k), k = 1, ..., n (например, для n = 7).


```python
print("Введите число: ")
n = int(input())

x = np.arange(-10, 10, 0.01)

for i in range(1, n+1):
    plt.plot(x, sc.stats.norm.pdf(x, 0, i))
```

2. Моделировать 1000 реализаций с.в. X ∼ Bin(100, 0.04) и с.в. Y ∼ Poiss(4). Построить столбцовые
диаграммы отдельно и на одном графике, сравнить. То же для X ∼ Bin(1000, 0.004) и X ∼
Bin(10, 0.4).

```python
bin_arr = []
poiss_arr = []
for i in range(1000):
    bin_arr.append(np.random.binomial(100, 0.04))
    poiss_arr.append(np.random.poisson(4))

    
plt.subplots_adjust(left = 0.01, right = 3)



plt.subplot(1, 3, 1)
sns.histplot(data = bin_arr, color = "#18a7b5", discrete=True, alpha = 0.5)
sns.histplot(data = poiss_arr, color = "#78866b", discrete=True, alpha = 0.5)
plt.ylabel('Count')

bin_arr.clear()
for i in range(1000):
    bin_arr.append(np.random.binomial(1000, 0.004))

plt.subplot(1, 3, 2)
sns.histplot(data = bin_arr, color = "#18a7b5", discrete=True, alpha = 0.5)
sns.histplot(data = poiss_arr, color = "#78866b", discrete=True, alpha = 0.5)
plt.ylabel('Count')

bin_arr.clear()
for i in range(1000):
    bin_arr.append(np.random.binomial(10, 0.4))

plt.subplot(1, 3, 3) 
sns.histplot(data = bin_arr, color = "#18a7b5", discrete=True, alpha = 0.5)
sns.histplot(data = poiss_arr, color = "#78866b", discrete=True, alpha = 0.5)
plt.ylabel('Count')
```

8. Для массива данных "titanic" сравнить выживших и невыживших пассажиров с помощью парал-
лельных координат, используя столбцы survived, pclass, who (нужно будет присвоить числовые
значения), age. (Не стоит использовать весь массив, лучше выбрать случайным образом 30 строк).

```python
data = sns.load_dataset('titanic')
print(data['age'].max(), data['age'].min())
```

```python
data
```

```python
from pandas.plotting import parallel_coordinates

def genfunc(x):
    if x['who'] == 'man':
        ret = 0
    elif x['who'] == 'woman':
        ret = 40
    else:
        ret = 80
    
    return ret


new_data = data.sample(100)
new_data['who_val'] = new_data.apply(genfunc, axis=1)
plt.figure(figsize=(10, 10))
new_data['pclass'] = (new_data['pclass'] - 1) * 40
new_data = new_data[['who_val', 'pclass', 'survived', 'age']]
new_data = new_data[new_data['age'] > 0]

parallel_coordinates(new_data, 'survived', color=('#EE9086', '#007DFF', '#C7F464'), alpha = 0.5)
```
