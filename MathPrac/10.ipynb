```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy as sc
import seaborn as sns

from statsmodels.distributions.empirical_distribution import ECDF

matplotlib.style.use('ggplot')
```

```python
#fixed - параметр, с которым генерируется выборка
def plotting(n, fixed):
    sample = sc.stats.bernoulli.rvs(fixed, size=n)
    thetta = np.arange(0, 1, 0.01)
    param = np.sum(sample)
    plt.plot(thetta, sc.stats.beta.pdf(thetta, param+1, n-param+1), label=str(n))
    plt.legend(loc='best')
    
N = [5, 10, 20, 50, 100]
```

```python
for n in N:
    plotting(n, 1/2)
```

```python
for n in N:
    plotting(n, 1/3)
```

```python
for n in N:
    thetta = np.arange(0, 1, 0.01)
    param = 9*n/10
    plt.plot(thetta, sc.stats.beta.pdf(thetta, param+1, n-param+1), label=str(n))
    
plt.legend(loc='best')
```

```python
def plotting(n, m, s, fixed):
    sample = sc.stats.norm.rvs(fixed, 1, size=n)
    mean = (m/(s**2) + np.sum(sample))/(1/s**2 + n)
    variance = 1/(1/s**2 + n)
    thetta = np.arange(-2*variance**0.5+mean, mean+2*variance**0.5, 0.001)
    plt.plot(thetta, sc.stats.norm.pdf(thetta, mean, variance**0.5), label=str(m)+' '+str(s))
    plt.legend(loc='best')
    
M = [-1, 0, 1]
S = [0.1, 1.5]
```

```python
fixed = 0.232454
for m in M:
    for s in S:
        plotting(1000, m, s, fixed)
```

```python
fixed = 0.00019043
for m in M:
    for s in S:
        plotting(1000, m, s, fixed)
```
