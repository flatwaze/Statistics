    по русскому и "нет" иначе (можно true, false),
    4. создать новый датасет, содержащий только абитуриентов со средним баллом больше 95.

```python
data = pd.read_csv("Priem.csv")
data.iloc[6:20]
```

```python
data.iloc[6:20, 2:4]
```

```python
def func(x):
    if x["ЕГЭ...рус.яз."] > x["ЕГЭ...матем."]:
        return 'Да'
    else:
        return 'Нет'

data['Новый столбец'] = data.apply(func, axis=1)

data
```

```python
ndata = data[(data['ЕГЭ...матем.'] + data['ЕГЭ...рус.яз.']) > 190]
ndata
```

3. Добавим группировку (groupby) и агрегирующие функции (agg) в Python или аналогичные функции
пакета dplyr в R:
    1. вывести средние баллы девочек и мальчиков
    2. вывести средние баллы, имеющих аттестат с отличием и нет,
    3. вывести средние баллы по 4 категориям, получающимся при сочетании двух предыдущих пунктов,
    4. найти число студентов с баллом ЕГЭ по математике, лежащим в заданном диапазоне,
    5. создать датасет с количествами студентов, баллы которых лежат в заданных диапазонах (<70,
    70-80 и т.д.).

```python
print("Средние баллы девочек и мальчиков: ")
data.iloc [:, [0,1]].groupby('Пол').mean()
```

```python
print("По аттестату: ")
data.iloc [:, [1, 4]].groupby('Аттестат.с.отличием.оценка').mean()
```

```python
print("По обоим вышеперечисленным параметрам: ")
data.iloc[:, [0, 1, 4]].groupby(['Аттестат.с.отличием.оценка', 'Пол']).mean()
```
